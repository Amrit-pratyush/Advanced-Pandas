Q3 Customer Purchase Analysis

Identify the top 5 customers based on total sales.

Hint: Use GroupBy with aggregation.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Advanced_pandas_sales_dataset.csv')

# 1. Group by Customer and aggregate total Sales
customer_sales = df.groupby('Customer')['Sales'].sum().reset_index()

# 2. Sort customers by Sales in descending order and grab the top 5
top_5_customers = customer_sales.sort_values(by='Sales', ascending=False).head(5)

print("--- TOP 5 CUSTOMERS BY TOTAL SALES ---")
print(top_5_customers.to_string(index=False, formatters={'Sales': '${:,.2f}'.format}))
```

Q4 Regional Sales Performance

Calculate the total sales generated by each region.

Then identify:


*   Which region generated the highest revenue?



```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Advanced_pandas_sales_dataset.csv')

# 1. Calculate the total sales generated by each region
region_sales = df.groupby('Region')['Sales'].sum()

print("--- TOTAL SALES BY REGION ---")
for region, sales in region_sales.items():
    print(f"{region:<10} : ${sales:,.2f}")

# 2. Identify the region with the highest revenue using idxmax()
highest_revenue_region = region_sales.idxmax()
highest_revenue_value = region_sales.max()

print("\n--- REVENUE LEADER ---")
print(f"The region that generated the highest revenue is the {highest_revenue_region} region with ${highest_revenue_value:,.2f}.")
```

Q5 Product Performance Analysis
1. Calculate the total quantity sold for each product.
2. Which product appears to be the most popular?

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Advanced_pandas_sales_dataset.csv')

# 1. Calculate the total quantity sold for each product
product_qty = df.groupby('Product')['Quantity'].sum().reset_index()

# 2. Sort the products by quantity in descending order to identify the most popular item
product_qty_sorted = product_qty.sort_values(by='Quantity', ascending=False)

print("--- TOTAL QUANTITY SOLD BY PRODUCT ---")
print(product_qty_sorted.to_string(index=False))

# 3. Identify the most popular product
most_popular_product = product_qty_sorted.iloc[0]['Product']
most_popular_quantity = product_qty_sorted.iloc[0]['Quantity']

print("\n--- POPULARITY SUMMARY ---")
print(f"The most popular product is '{most_popular_product}' with a total of {most_popular_quantity} units sold.")
```

Q6 Multi-Level Grouping

Its weakest performance is in the East region ($14,122.00) and its strongest is in the North region ($26,135.00), showing that office products have low market penetration or lower average order values.

*   Accessories Regional Patterns: Accessories perform exceptionally well in the South region ($80,914.00), outperforming the North, West, and East branches for that specific category.







Q7 Pivot Table Creation

Create a pivot table showing:

*   Rows → Region
*   Columns → Product
*   Values → Total Sales

Which region sells the most laptops?

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Advanced_pandas_sales_dataset.csv')

# Create the pivot table using pivot_table() with aggfunc='sum' to aggregate sales
sales_pivot = df.pivot_table(index='Region', columns='Product', values='Sales', aggfunc='sum')

print("--- REGIONAL PRODUCT SALES PIVOT TABLE ---")
print(sales_pivot)

# Identify which region sells the most laptops
highest_laptop_region = sales_pivot['Laptop'].idxmax()
highest_laptop_sales = sales_pivot['Laptop'].max()

print("\n--- LAPTOP SALES OVERVIEW ---")
print(f"The region that sells the most Laptops is the '{highest_laptop_region}' region with a total of ${highest_laptop_sales:,.2f} in sales.")
```

Q8 Profitability Analysis

*   Calculate the average profit for each product category.
*   Which category is most profitable?


```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Advanced_pandas_sales_dataset.csv')

# 1. Calculate the average profit for each product category
category_profit = df.groupby('Category')['Profit'].mean().reset_index()

# 2. Sort to easily find the most profitable category
category_profit_sorted = category_profit.sort_values(by='Profit', ascending=False)

print("--- AVERAGE PROFIT BY PRODUCT CATEGORY ---")
print(category_profit_sorted.to_string(index=False, formatters={'Profit': '${:,.2f}'.format}))

# 3. Identify the most profitable category
most_profitable_category = category_profit_sorted.iloc[0]['Category']
highest_average_profit = category_profit_sorted.iloc[0]['Profit']

print("\n--- PROFITABILITY LEADER ---")
print(f"The most profitable category on average is '{most_profitable_category}' with an average profit of ${highest_average_profit:,.2f} per order.")
```

Q9 Discount Impact Study
Analyze the relationship between discount and profit.

Tasks:
1. Calculate the average profit for each discount level
2. Identify whether higher discounts lead to lower profit

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Advanced_pandas_sales_dataset.csv')

# 1. Calculate the average profit for each unique discount level
discount_impact = df.groupby('Discount')['Profit'].mean().reset_index()

print("--- AVERAGE PROFIT BY DISCOUNT LEVEL ---")
print(discount_impact.to_string(index=False, formatters={'Discount': '{:.2%}'.format, 'Profit': '${:,.2f}'.format}))
```

$\text{Analytical Conclusion}$

$\text{Do higher discounts lead to lower profit?}$ Yes, absolutely.

There is a clear, inverse linear relationship observed here: as the discount level scales up from $0.00\%$ to $20.00\%$, the average net profit per transaction drops dramatically from $751.21 down to just $90.03. This reveals that aggressive discount policies are severely compressing product margins and cutting into net profitability.

Q10 End-to-End Data Wrangling Task

Perform the following workflow:
1. Filter the dataset where Sales > 3000
2. Group data by Region and Product
3. Calculate total sales
4. Reshape the result into a pivot table

Explain the business insights obtained from this analysis.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Advanced_pandas_sales_dataset.csv')

# Step 1: Filter the dataset where Sales > 3000
filtered_df = df[df['Sales'] > 3000]

# Steps 2 & 3: Group data by Region and Product, and calculate total sales
grouped_sales = filtered_df.groupby(['Region', 'Product'])['Sales'].sum().reset_index()

# Step 4: Reshape the result into a pivot table
# This maps out high-value transaction distributions seamlessly
wrangled_pivot = filtered_df.pivot_table(index='Region', columns='Product', values='Sales', aggfunc='sum')

print("--- END-TO-END WRANGLED PIVOT TABLE (SALES > 3000) ---")
print(wrangled_pivot.fillna(0)) # Using fillna(0) ensures clean visualization if any combo has zero records
```

$\text{Business Insights Obtained}$

$\text{Premium Laptop Hub in the East:}$ Even when filtering down exclusively to large bulk or premium-tier orders (Sales $> 3000$), Laptops in the East region bring in an exceptional total revenue of $44,096.00. This highlights the East region as a critical geographic market for high-ticket corporate enterprise or premium computer purchases.

$\text{Accessories Driving High Ticket Baskets:}$ Interestingly, lower-cost items like Keyboards in the West (\$33,380.00) and Mice in the South (\$26,928.00) show prominent transaction figures in the high-value category. Since individual components are cheap, this reveals a high volume of bulk order behavior (e.g., procurement departments fitting out entire office floors at once) rather than lone consumer shopping.

$\text{Underperforming High-Value Categories:}$ Printers in the East ($3,696.00$) and Tablets in the North ($3,392.00$) are barely breaking past the minimum thresholds. Marketing and B2B sales teams should focus promotional bundles or account targeting strategies to improve wholesale corporate pipelines for these specific regional product segments.
